- If the variance–covariance matrix of the predictors differs across the two groups then the log odds are a function of x , x2 and $ x_ix_j (i, j = 1,..., p; i \ne j) $. A quadratic term in xi is needed as a predictor if the variance of xi differs across the two values of y. The product term $x_i x_j$ is needed as a predictor if the covariance of $x_i$ and $x_j$ differs across the two values of y (i.e., if the regression of $x_i$ on $x_j$ (or vice versa) has a different slope for the two values of y.)
- If the densities $f(x|Y = j), j = 0,1$ are skewed the log odds can depend on both x and log(x). It does, for example, for the gamma distribution. Cook and Weisberg (1999b, p. 501) give the following advice:
    - *"When conducting a binary regression with a skewed predictor, it is often easiest to assess the need for x and log(x) by including them both in the model so that their relative contributions can be assessed directly."*
- Boxplots split by $y \in \{0, 1\}$ to look for skewed data and differing mean/variance

#### When to use this model (Assumptions of Logistic Regression)
- You believe your data doesn't have much inter-dependence
- You have labeled data and your problem is a classification problem
- Features are independent of one another (AKA no multicollinearity)
- Sample size is adequate – Rule of thumb: 50 records per predictor
- Only the meaningful variables should be included. (No automatic feature selection)
- You have linearity of independent variables and log odds results. AKA your independent/dependent data doesn't have a piecewise relationship.
- Your observations are independent (no time series)

#### Notes
- With logistic regression, you are responsible for giving it relevent, "clean" data

- A confusing naming convention "glitch" is that in machine learning, "regression" normally refers to numerical predictions, ie stock market, whilst "classification" normally refers to catagorical predictions, ie has_cancer. However, *logistic regression* is a classification algorithm. 
- If statiticians ever talk about $log()$, they most likely certainly mean natural log, $ln()$. We will follow that convention here.
- Another convention with stataticians is to write $exp\{x\}$ to mean $e^x$

#### How it works
- *Hold on, this is a rollercoaster. If you get lost, just continue with the notebook. It's not absolutely neccessary you know this.*
- Linear regression (of which logistic regression is a class of) means that your outcome can be described as a __continous, linear__ function of your features.
    - This is in contradiction to other models, like a decision tree which describes the outcome in more of a piecewise function.
- y = b0 + b1x1 + b2x2 + ... + bnxn is the traditional __linear regression model__ where xi is the ith feature of your dataset. b0 is the y-intercept and {bi | 1 ≤ i ≤ n} are your feature coeffients. This model can be interpreted as "a one unit increase in xi corresponds to a bi increase in y". Bi is the *predicted average* increase in y for a one unit increase in xi, so more accurately we could say,  "__Our model predicts__ that a one unit increase in xi corresponds, __on average__, to a bi increase in y".
- We can use matrix notation to summarize b0 + b1x1 + b2x2 + ... + bnxn as B\*X where B is a vector of coeffecients and X is a matrix of your feature values.
- log(y) = B\*X is the traditional __logistic regression model__. This model can be interpreted as "a one unit increase in xi corresponds to a Bi __log odds__ increase in y". That is, B\*X has a linear (straight line) relationship with the __log odds__. Often, the model gets rewritten to y = exp{B\*X}. In this case, we interpret the model as "a one unit increase in xi corresponds to an exp{bi} increase in the odds of y". 
- __Odds are different than probability__. Odds is success/failures. The domain for odds is \[0, inf). Probability is success/total with total being success+failures. The domain for probability is \[0, 1\]. P = odds/(1+odds)

#### A few points for comparison:
![table comparison](http://www.statisticshowto.com/wp-content/uploads/2017/12/conversions.png)
###### The important one is .5 probability (think "50/50 chance") means 1 odds (think "1 to 1 odds") and 0 log odds
How to convert between them:
- Take glm output coefficient (logit)
- compute e-function on the logit using exp() “de-logarithimize” (you'll get odds then)
- convert odds to probability using this formula prob = odds / (1 + odds).

```python
# We should start to get an idea from these boxplots about
# which features will play a role in determining our outcome.
# Which features have the most signifigant difference in mean?
def show_boxplots(predictors, targets):
    n_features = len(list(predictors))
    fig, axs = plt.subplots(nrows=math.ceil(n_features / 3), 
                            ncols=math.floor(n_features / 3),
                            figsize=(10,10))
    for i, col_name in enumerate(predictors):
        sns.boxplot(x='vote', y=col_name, data=pd.concat([predictors, targets], axis=1), ax=axs[i // 3, i % 3])
    plt.show()
```

```python
show_boxplots(X, y)
```

## Naive Fit

```python
# Build a logistic regression model and score it
def get_model_score(predictors, targets):
    X_train, X_test, y_train, y_test = train_test_split(predictors, targets, test_size=0.3, stratify=targets, random_state=123456)
    log_reg = LogisticRegression()
    model = log_reg.fit(X_train, y_train)
    predicted = model.predict(X_test)
    return statistics.mean(cross_val_score(model, predictors, targets, cv=5)), model
```

```python
score, model = get_model_score(X, y)
print(f"Current score is {round(score, 3)}.")
```

```python
def get_model_metrics(predictors, targets):
    X_train, X_test, y_train, y_test = train_test_split(predictors, targets, test_size=0.3, stratify=targets, random_state=123456)
    log_reg = LogisticRegression()
    model = log_reg.fit(X_train, y_train)
    predicted = model.predict(X_test)
    print(classification_report(y_test, predicted))
    sns.heatmap(confusion_matrix(y_test, predicted), annot=True, fmt='g')
```

```python
def cross_feature_regression(data, feature1, feature2):
    g = sns.FacetGrid(data, hue='vote', size=5)
    g.map(plt.scatter, feature1, feature2, s=50, alpha=.7, linewidth=.5, edgecolor="white")
    g.map(sns.regplot, feature1, feature2, ci=None, robust=1, x_jitter=.1, y_jitter=.1, marker="o", scatter_kws={"alpha":.3})
    g.add_legend()
```
